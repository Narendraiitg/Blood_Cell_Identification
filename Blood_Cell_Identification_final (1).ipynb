```python
import numpy as np                      #a popular python library
import cv2                              #open cv has a function to read videos, which is cv2
import matplotlib.pyplot as plt         # python packages it gives MATLAB like way of plotting
# read original image
image = cv2.imread("C:/Users/Rahimul/Desktop/c120.png") # read an image from a file into an array
# Output Images
plt.imshow(image,cmap='gray')           #creates an image from a 2D numpy array
plt.title('Original Image')             #image display with title
plt.figure()                            # ploting

   # convet to gray scale image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  #
# Output Images
plt.imshow(gray,cmap='gray')
plt.title('Grayscale Image')
plt.figure()

   # apply median filter for smoothning
blurM = cv2.medianBlur(gray, 5)
# Output Images
plt.imshow(blurM,cmap='gray')
plt.title('Median filtered output')
plt.figure()

    # apply gaussian filter for smoothning
blurG = cv2.GaussianBlur(gray, (9, 9), 0)
# Output Images
plt.imshow(blurG,cmap='gray')
plt.title('Gaussian output')
plt.figure()

# histogram equalization
histoNorm = cv2.equalizeHist(gray)
# Output Images
plt.imshow(histoNorm,cmap='gray')
plt.title('Histogram Equalized output')
plt.figure()


    # create a CLAHE object for contrast Limited Adaptive Histogram Equalization (CLA
clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize=(8, 8))
claheNorm = clahe.apply(gray)
# Output Images
plt.imshow(claheNorm,cmap='gray')
           
    # contrast stretching
# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, s1, r2, s2):
    if (0 <= pix and pix <= r1):
        return (s1 / r1) * pix
    elif (r1 < pix and pix <= r2):
        return ((s2 - s1) / (r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2) / (255 - r2)) * (pix - r2) + s2
 

 # Define parameters.
 
 
r1 = 70
s1 = 0
r2 = 200
s2 = 255
           
           
           # Vectorize the function to apply it to each value in the Numpy array. 
pixelVal_vec = np.vectorize(pixelVal) 
# Apply contrast stretching. 
contrast_stretched = pixelVal_vec(gray, r1, s1, r2, s2) 
contrast_stretched_blurM = pixelVal_vec(blurM, r1, s1, r2, s2)
plt.imshow(contrast_stretched_blurM)
plt.show()
cv2.imwrite('Final_Output_colour.png', contrast_stretched_blurM)
           
           
           
           # edge detection using canny edge detector 
edge = cv2.Canny(gray, 100, 200) 
edgeG = cv2.Canny(blurG, 100, 200) 
edgeM = cv2.Canny(blurM, 100, 200) 
cv2.imwrite('Final_Output.png', edgeM)
# Output Images
plt.imshow(edge,cmap='gray')

        # edge detection using canny edge detector 
edge = cv2.Canny(gray, 100, 200) 
edgeG = cv2.Canny(blurG, 100, 200) 
edgeM = cv2.Canny(blurM, 100, 200) 
cv2.imwrite('Final_Output.png', edgeM)
# Output Images
plt.imshow(edge,cmap='gray')
plt.figure()
# Output Images
plt.imshow(edgeG,cmap='gray')
# Output Images
plt.imshow(edgeM,cmap='gray')
# Output Images
plt.imshow(edgeG,cmap='gray')
plt.figure()
```
